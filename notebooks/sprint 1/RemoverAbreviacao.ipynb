```python
norm = Normaliser(tokenizer='readable')
chats_clients = pd.read_csv('/content/drive/Shareddrives/grupo3moshi/chats_clients.csv')
chats_clients.head()
```

```python
def correct_grammar(text):
  return norm.normalise(text)
```

```python
chats_clients['Pergunta_corrected_grammar'] = chats_clients['Pergunta'].apply(correct_grammar)
chats_clients['Resposta_corrected_grammar'] = chats_clients['Resposta'].apply(correct_grammar)
```
